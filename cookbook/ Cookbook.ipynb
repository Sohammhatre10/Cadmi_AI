 # Cookbook
 
 A Jupyter notebook is an interactive web-based environment that allows users to create and share documents containing live code, equations, visualizations, and narrative text.
 
 Link of Jupyter website: https://jupyter.org 
 Link of Jupyter Notebook installation: https://jupyter.org/install 
 
# Jupyter Notebook
1. Install the classic Jupyter Notebook with:

pip install notebook

2. To run the notebook:

jupyter notebook

3. You can also access Jupyter Notebook through anaconda. 

Link to documentation: https://docs.conda.io/en/latest/

1. File

- New Notebook: Create a new notebook and click on Python 3.

- Open: Open files on your browser/computer.

- Make a Copy: Copies the notebook (opened in a separate window).

- Save as: Can be used to save document in a separate folder or name.

- Rename: Renaming the notebook.

- Save and Checkpoint: Saving your notebook locally (but remember to commit).

- Revert to checkpoint: Go back to previous version.

- Print Review: To review how to print your notebook.

- Download as: Options to download in different versions.

- Close and hault: close notebook.


2. Edit

- Undo/Redo: Allows you to revert or reapply recent changes made to the notebook.

- Cut/Copy/Paste: Lets you cut or copy selected cells and paste them in the desired location, facilitating easy organization of your work.

- Insert Cells: You can insert new code or markdown cells above or below the currently selected cell.

- Delete Cells: This option removes the selected cell from the notebook.

- Merge Cells: This option combines two or more selected cells into a single cell. 

- Move Cell: This option allows you to change the order of your cells by moving a selected cell up or down in the notebook.
 
3. View

Toggle Options

- Toggle Header: Shows or hides the notebook's header, allowing you to focus on the content without distractions.
    
- Toggle Toolbar: Displays or hides the toolbar at the top of the notebook, which contains quick access buttons for common actions like saving, adding cells, and running code.
    
- Toggle Line Numbers: Shows or hides line numbers in code cells, which can help with debugging and reviewing code.

4. Cell Toolbar

- Cell Toolbar: This feature provides various options to display additional tools for managing cells. When you select a cell toolbar option, it appears on the right side of the cell and can include:

  - Tags: Allows you to tag cells for easy identification or organization.
        
  - Slideshow: Enables you to prepare the notebook for presentation by defining which cells to include in a slideshow.
        
 - Edit Metadata: Lets you edit the metadata associated with a cell, which can be useful for more advanced configurations.
        

5. Insert

- Insert Cell Above/Below: Adds a new code or markdown cell directly above or below the currently selected cell, helping you easily expand your notebookâ€™s content.

6. Cell

- Run Cells: Executes the code in the selected cell(s).

- Insert Cell Below: Quickly adds a new cell below the current one.

- Delete Cells: Removes the selected cell(s) from the notebook.

- Change Cell Type: Allows you to switch a cell between code and markdown formats, which helps in organizing your content.

7. Kernel 

- Restart: Stops and restarts the kernel, which is useful if you want to clear all variables and outputs.

- Interrupt: Halts the execution of the current code cell, useful for stopping long-running processes.

- Restart & Run All: Restarts the kernel and runs all cells in the notebook sequentially.


8. Widgets

- It provides options to manage and create interactive widgets, such as sliders, buttons, and dropdowns. These widgets can be used to enhance the user experience by allowing dynamic input and interaction with code.

9. Help

- User Interface Tour: Provides a guided tour of the Jupyter Notebook interface.

- Keyboard Shortcuts: Lists available keyboard shortcuts to enhance navigation and productivity.

- Documentation: Links to the official Jupyter documentation for further reading and support.
